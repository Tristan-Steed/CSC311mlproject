### Importing libraries

```python
import numpy as np
import matplotlib as pt
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from pandas import DataFrame
```

### Importing Dataset.

```python
data = np.genfromtxt("mnist_data_selection.csv",delimiter=",", dtype="int")
# n is for last column of the dataset which is y
n = data.shape[1] -1

display(pd.DataFrame(data).head(10))
```

### Splitting data 

```python
X = np.array(data[:,0:n])
y = np.array(data[:,n:n+1])

display(pd.DataFrame(X).head(10))
display(pd.DataFrame(y).head(10))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2)

training_data = np.hstack((X_train, y_train))
testing_data = np.hstack((X_test, y_test))

train = DataFrame(training_data)
test = DataFrame(testing_data)

train.to_csv("group2_tr.csv", index= False, header = None)
test.to_csv("group2_te.csv", index = False, header = None)

```
